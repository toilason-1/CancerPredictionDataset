# PERCENTAGE OF CANCEROUS PATIENTS AMONG LOW BMI AND AND HIGH BMI

```python
low_bmi_patients = df[df['BMI'] < 25]
high_bmi_patients = df[df['BMI'] >= 25]

# Step 2: Count cancerous patients (Diagnosis = 1) in each BMI category
cancerous_low_bmi = low_bmi_patients[low_bmi_patients['Diagnosis'] == 1].shape[0]
cancerous_high_bmi = high_bmi_patients[high_bmi_patients['Diagnosis'] == 1].shape[0]

# Step 3: Total number of patients in each BMI category
total_low_bmi = low_bmi_patients.shape[0]
total_high_bmi = high_bmi_patients.shape[0]

# Step 4: Calculate percentages of cancerous patients
percentage_cancerous_low_bmi = (cancerous_low_bmi / total_low_bmi) * 100 if total_low_bmi > 0 else 0
percentage_cancerous_high_bmi = (cancerous_high_bmi / total_high_bmi) * 100 if total_high_bmi > 0 else 0

# Step 5: Prepare data for the bar chart
labels = ['Low BMI (<25)', 'High BMI (>=25)']
percentages = [percentage_cancerous_low_bmi, percentage_cancerous_high_bmi]

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(labels, percentages, color=['lightblue', 'lightcoral'])

# Add title and labels
plt.title('Percentage of Cancerous Patients among Low and High BMI Groups')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)  # Set y-axis limit to 100%
plt.grid(axis='y')  # Add gridlines for better readability

# Show the percentage values on the bars
for i, v in enumerate(percentages):
    plt.text(i, v + 1, f'{v:.1f}%', ha='center', fontweight='bold')

plt.show()
```

# scatter plot of cancer patient by genetic risk

```python
total_records_by_risk = df.groupby('GeneticRisk').size().reset_index(name='TotalRecords')
cancer_count_by_risk = df[df['Diagnosis'] == 1].groupby('GeneticRisk').size().reset_index(name='CancerCount')

# Merge the two DataFrames on GeneticRisk
merged_data = pd.merge(total_records_by_risk, cancer_count_by_risk, on='GeneticRisk', how='left').fillna(0)

# Calculate the percentage of cancer patients
merged_data['CancerPercentage'] = (merged_data['CancerCount'] / merged_data['TotalRecords']) * 100

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='GeneticRisk', y='CancerPercentage', data=merged_data, s=100, color='blue')

# Add title and labels
plt.title('Percentage of Cancer Patients by Genetic Risk')
plt.xlabel('Genetic Risk (0: No Risk, 1: Medium Risk Risk,2:High Risk)')
plt.ylabel('Percentage of Cancer Patients (%)')
plt.xticks([0, 1,2])  # Set x-ticks to match the genetic risk values
plt.grid(True)
plt.show()
```

PIE CHART OF PATIENTS WITH CANCER HISTORY

```python
patients_with_cancer_history = df[df['CancerHistory'] == 1]

# Step 2: Count diagnosed (Diagnosis = 1) and not diagnosed (Diagnosis = 0) patients
diagnosis_counts = patients_with_cancer_history['Diagnosis'].value_counts()
print(diagnosis_counts)

# Step 3: Calculate the percentages
percentages = (diagnosis_counts / diagnosis_counts.sum()) * 100

# Step 4: Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(percentages, labels=['Diagnosed (1)', 'Not Diagnosed (0)'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title('Proportion of Diagnosed and Not Diagnosed Patients with Cancer History')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

83.3 % of data with a cancer history is diagnosed with cancer

# PIE CHART OF SMOKER

```python
smokers = df[df['Smoking'] == 1]

# Step 2: Count cancer patients among smokers
cancer_patients = smokers[smokers['Diagnosis'] == 1].shape[0]

# Step 3: Total number of smokers
total_smokers = smokers.shape[0]

# Step 4: Calculate percentage of smokers who have cancer
if total_smokers > 0:  # To avoid division by zero
    percentage_with_cancer = (cancer_patients / total_smokers) * 100
else:
    percentage_with_cancer = 0

# Step 5: Prepare data for the pie chart
labels = ['Smokers with Cancer', 'Smokers without Cancer']
sizes = [percentage_with_cancer, 100 - percentage_with_cancer]  # 100% minus the cancer percentage

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue'])
plt.title('Percentage of Smokers with and without Cancer')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

55.2 percentage of people who smoke is diagnosed with cancer

```python
non_smokers = df[df['Smoking'] == 0]

# Step 2: Count cancer patients among non-smokers
cancer_patients = non_smokers[non_smokers['Diagnosis'] == 1].shape[0]

# Step 3: Total number of non-smokers
total_non_smokers = non_smokers.shape[0]

# Step 4: Calculate percentage of non-smokers who have cancer
if total_non_smokers > 0:  # To avoid division by zero
    percentage_with_cancer = (cancer_patients / total_non_smokers) * 100
else:
    percentage_with_cancer = 0

# Step 5: Prepare data for the pie chart
labels = ['Non-Smokers with Cancer', 'Non-Smokers without Cancer']
sizes = [percentage_with_cancer, 100 - percentage_with_cancer]  # 100% minus the cancer percentage

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue'])
plt.title('Percentage of Non-Smokers with and without Cancer')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()
```

```python
smokers_with_cancer = df[(df['Smoking'] == 1) & (df['Diagnosis'] == 1)].shape[0]
non_smokers_with_cancer = df[(df['Smoking'] == 0) & (df['Diagnosis'] == 1)].shape[0]

# Step 2: Total number of smokers and non-smokers
total_smokers = df[df['Smoking'] == 1].shape[0]
total_non_smokers = df[df['Smoking'] == 0].shape[0]

# Step 3: Calculate percentages
percentage_smokers_with_cancer = (smokers_with_cancer / total_smokers) * 100 if total_smokers > 0 else 0
percentage_non_smokers_with_cancer = (non_smokers_with_cancer / total_non_smokers) * 100 if total_non_smokers > 0 else 0

# Step 4: Prepare data for the bar chart
labels = ['Smokers with Cancer', 'Non-Smokers with Cancer']
percentages = [percentage_smokers_with_cancer, percentage_non_smokers_with_cancer]

# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(labels, percentages, color=['lightcoral', 'lightblue'])

# Add title and labels
plt.title('Percentage of Smokers and Non-Smokers with Cancer')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)  # Set y-axis limit to 100%
plt.grid(axis='y')  # Add gridlines for better readability

# Show the percentage values on the bars
for i, v in enumerate(percentages):
    plt.text(i, v + 1, f'{v:.1f}%', ha='center', fontweight='bold')

plt.show()
```
